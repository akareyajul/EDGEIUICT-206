```python
# Generating Heatmap
plt.figure(figsize=(10,10))
sns.heatmap(co_matrix, annot=True, linecolor='white', cbar=True, cbar_kws=None, cbar_ax=None)
plt.title("Correlation Matrix")
```

```python
# Plot a histogram of total_rooms and  its distribution
plt.hist(data['total_rooms'], bins=20)
plt.xlabel('total_rooms')
plt.ylabel('Frequency')
plt.title('Histogram of total_rooms')
plt.show()
```

```python
# Handle Outlier and Preprocessing
Q1 = data['median_house_value'].quantile(0.25)
Q3 = data['median_house_value'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier = data[(data['median_house_value'] < lower_bound) & (data['median_house_value'] > upper_bound)]
len(outlier)
```
